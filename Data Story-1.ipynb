```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
%matplotlib inline
```

```python
pd.set_option('display.max_columns', 26)
df = pd.read_csv('ExerciseDataSet.csv')
df = df.drop(df.columns[-1], axis=1)
df
```

```python
df.count()
```

#### Table of missing data:
Above is a chart that shows the amount of values present in each column of data. For the column to be full and not be missing data, the count number should be 40. This is the case for most columns except for VO2MAX which has 39, meaning it is missing one value, and VO2MAX2 which has 36, meaning it is missing 4 values.

```python
sns.heatmap(df.isnull(),yticklabels=False,cbar=False,cmap='viridis')
plt.title('Areas of Missing Data in the Data Set')
```

#### Heatmap of Missing Data:
Since the data set is so large, I figured it would be a good idea to incorporate a heatmap that shows exactly on the table where the data is missing. The yellow rectangle represents the missing data cells.

```python
sns.set_style('whitegrid')

fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(16, 3))
sns.countplot(x='HADSD', hue='Group', data=df, ax=ax1)
ax1.set_xlim(0, 21)
ax1.set_ylim(0, 5)
ax1.set_xticks(range(0, 21, 1))
ax1.set_xticklabels(range(0, 21, 1))

sns.countplot(x='HADSD2', hue='Group', data=df, ax=ax2)
ax2.set_xlim(0, 21)
ax2.set_ylim(0, 5)
ax2.set_xticks(range(0, 21, 1))
ax2.set_xticklabels(range(0, 21, 1))

plt.show()
```

```python
sns.set_style('whitegrid')

fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(16, 3))
sns.countplot(x='HADSA', hue='Group', data=df, ax=ax1)
ax1.set_xlim(0, 21)
ax1.set_ylim(0, 5)
ax1.set_xticks(range(0, 21, 1))
ax1.set_xticklabels(range(0, 21, 1))

sns.countplot(x='HADSA2', hue='Group', data=df, ax=ax2)
ax2.set_xlim(0, 21)
ax2.set_ylim(0, 5)
ax2.set_xticks(range(0, 21, 1))
ax2.set_xticklabels(range(0, 21, 1))

plt.show()
```

### Bar Graph explanation and Elaboration
In order to get a good visual representation of how the levels of depression varied between the two groups before and after the exercise intervention, I figured a bar graph would suffice. The four graphs contain blue bars which are the bars of the control group and orange bars for the exercise group. The x-Axis represents the value that the people ranked on the HADS scale. 0-7 on the scale represents normal levels of depression and anxiety. 8-10 represents borderline abnormal amounts which is potentially a case. 11-21 represents an abnormal amount which is a case for depression or anxiety. The y-Axis represents the frequency in which each value came up. I chose to visualize the HADS levels since they were a prominent focus in this study. In the first graph which assesses depression levels, we can see that levels of depression tended to drop after the intervention for the exercise group whereas the control group seemed to increase overall. In the second set of graphs anxiety levels seemed to spread out after the intervention phase for both groups showing that there was little impact from the exercise.

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 3))

sns.set_style('darkgrid')
sns.scatterplot(x='HADSA', y='HADSD', hue='Group', data=df, ax=ax1, palette="deep")
ax1.set_xticks(range(0, 21, 1))
ax1.set_xticklabels(range(0, 21, 1))
ax1.set_yticks(range(0, 21, 1))
ax1.set_yticklabels(range(0, 21, 1))

sns.set_style('darkgrid')
sns.scatterplot(x='HADSA2', y='HADSD2', hue='Group', data=df, ax=ax2, palette="deep")
ax2.set_xticks(range(0, 21, 1))
ax2.set_xticklabels(range(0, 21, 1))
ax2.set_yticks(range(0, 21, 1))
ax2.set_yticklabels(range(0, 21, 1))

plt.show()

```

### Scatterplot Explanation and Elaboration:
After displaying the data on the bar graphs, I wanted to find a visualization that could combine both depression and anxiety scales into one concise set of graphs. The scatterplot was able to show all this data in a more digestible format. Above you can see a before exercise intervention and after intervention scatter plots. The blue dots represent the control group members just like the bar graphs and the orange represents the exercise group. These graphs make it much easier to see the effects of the exercise period. 

```python
sns.set_style('darkgrid')
sns.lmplot(x="HADSA", y="HADSD", hue="Group", data=df)
plt.xlim(0, 20)
plt.ylim(0, 20)
plt.title('Data Regression Before Exercise')
plt.show()
```

```python
sns.set_style('darkgrid')
sns.lmplot(x="HADSA2", y="HADSD2", hue="Group", data=df)
plt.xlim(0, 20)
plt.ylim(0, 20)
plt.title('Data Regression After Exercise')
plt.show()
```

### Regression Plot Explanation. 
Above are two scatterplots with regression lines on them to show the change in averages between the before and after period. I felt that this would help the audience get the best possible visualization in terms of visual clarity while also providing all the data of the data set.
